```python
%%time
def rw(a,b):
    c=b.lower()
    a=a.replace(b,c)
    return a

x=0

boseq=[]



File=open('broseq2.fsa','r')

for line in File:
    boseq.append(line.strip())
File.close()    

circle=0
test=True
while test:
    contig={}
    unnamed_D={}
    for seq in boseq :##first list 
        if seq[-20:] not in contig:##qu tou haishi quwei?
            contig[seq[-20:]]=0


    
    for contg in contig:
        unnamed_D[contg]=[]

        for XXX in boseq:

            if XXX.find(contg)!=-1:##wu fang hui de qu chu
                contig[contg]+=1
                boseq.remove(XXX)
                
                XXX=rw(XXX,contg)
                unnamed_D[contg].append(XXX)

    for contg in contig.keys():
        if contig[contg]==0:
            contig.pop(contg)
    
    for contg in unnamed_D.keys():
        if unnamed_D[contg]==[]:
            unnamed_D.pop(contg)


    import collections
    mycounts=collections.Counter(contig).most_common()

    #print unnamed_D
    x+=1
    contigFile=open("contigFile%s.txt"%x,"w")
    for i in mycounts:
        contigFile.write(i[0])
        contigFile.write('\n')
        for val in unnamed_D[i[0]]:            
                contigFile.write('\n')
                contigFile.write(val)
                contigFile.write('\n')
        contigFile.write('\n')
        contigFile.write('\n')
    contigFile.close()

    for key in unnamed_D.keys():##pin jie 
        if len(unnamed_D[key])>1:
            index1=[]
            index2=[]
            index3={}
            index4={}
            for seq in unnamed_D[key]:
                key=key.lower()
                site=seq.index(key)
                index1.append(site)
                index3[site]=seq
                lend=len(seq[site:])
                index2.append(lend)
                index4[lend]=seq
            #print index3
            #print index4
            begin=max(index1)
            end=max(index2)
            scaf=index3[begin][:begin]+index4[end][-end:]
            boseq.append(scaf)
        else :boseq.append(unnamed_D[key][0])

       
    for n,i in enumerate(boseq):
        boseq[n]=boseq[n].upper()

    
    circle+=1
        
    testL=[]
    for contg in contig.keys():
        testL.append(contig[contg])
    if max(testL)==1:
        test=False

print 'circle: ',circle
print 'you finally get ',len(boseq),'sequences'


result=open("result.txt","w")
for i in boseq:    
    result.write(i)
    result.write('\n')
    result.write('\n')
result.close()
##coverage
##homo_blast
for i in boseq:
    print len(i)
    



```
